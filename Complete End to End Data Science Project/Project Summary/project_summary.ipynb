**1.Check Missing Value:** There is no missing value in data

**2.Categorical Data Conversion:** To manage categorical data effectively, we employ frequency and manual encoding methods. This is particularly useful when dealing with features that encompass a high number of distinct labels.

* Mannual Encoding: Manual encoding, accomplished through the utilization of the "map" function, is an effective technique for managing categorical features. It involves associating labels with their corresponding frequencies, making it a valuable method in data preprocessing.

* Frequency Encoding: Frequency encoding is a method used to convert an original categorical variable into a numerical one by taking into account the frequency distribution of the data, specifically by analyzing the value counts.

**3.Outlier Handling** Certain features exhibit outliers, and to address this issue, we employ the Interquartile Range (IQR) method for outlier imputation. This approach is chosen because the data in these features does not follow a normal distribution.

* The Square Root Transformation is a commonly used standard transformation method. It is particularly suitable for count data that follows a Poisson distribution or small whole numbers. In this transformation, each data point is substituted with its square root. Negative values are made positive by adding a constant before applying the transformation.
* A Q-Q Plot, short for Quantile-Quantile Plot, is a graphical technique that compares two probability distributions by plotting their respective quantiles against each other.

* Standard Scaling, also known as standardization, is a technique used to rescale a feature. It operates under the assumption that the feature follows a normal distribution. It scales the feature to have a mean of 0 and a standard deviation of 1.
